```python
import pandas as pd

df = pd.read_csv("titanic.csv")
```

```python
df1 = df[(df["Sex"] == "female") & (df["Pclass"].isin([1,2]))]
df1["Age"] = df1["Age"].fillna(30)
```

```python
df1 = df1.sort_values("Fare", ascending=False).reset_index(drop=True)
```

```python
df1["fare_group"] = "high"
```

```python
df1.to_csv("female_upper_class.csv")
```
