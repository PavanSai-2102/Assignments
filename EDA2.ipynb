```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("daily_sales_records.csv")
```

# Histogram of units_sold (overall)

```python
def plot_units_histogram(df):
    sns.histplot(data = df, x = "units_sold", bins = 20)
    plt.xlabel("Units Sold")
    plt.ylabel("Count")
    plt.title("Distribution of Sold Units")
    plt.show()
```

# Histogram of units_sold by category

```python
def plot_units_histogram_hue(df):
    sns.histplot(data = df, x = "units_sold",hue = "category" ,bins = 20)
    plt.xlabel("Units Sold")
    plt.ylabel("Count")
    plt.title("Distribution of Sold Units")
    plt.show()
```

# Box plot of revenue by category

```python
def plot_revenue_boxplot(df):
    sns.boxplot(data= df, x = "category", y = "revenue")
    plt.xlabel("Category")
    plt.ylabel("Revenue")
    plt.title("Box plot of Revenue by Category")
    plt.show()
```

# Scatter plot of units_sold vs revenue

```python
def plot_scatter(df):
    sns.scatterplot(data= df, x = "units_sold", y = "revenue" , hue = "category")
    plt.xlabel("Units sold")
    plt.ylabel("Revenue")
    plt.title("Scatter Plot of Units sold vs Revenue by Category")
    plt.grid()
    plt.show()
```

# Outlier bounds

```python
def compute_outlier_bounds(df):
    q1 = df["revenue"].quantile(0.25)
    q3 = df["revenue"].quantile(0.75)
    IQR = q3 - q1
    Lower = q1-(1.5*IQR)
    upper = q3+(1.5*IQR)
    print(f"Revenue outlier bounds → Lower: {Lower}, Upper: {upper}")
```

```python
if __name__ == "__main__":
    plot_units_histogram(df)
    plot_units_histogram_hue(df)
    plot_revenue_boxplot(df)
    plot_scatter(df)
    compute_outlier_bounds(df)
```
