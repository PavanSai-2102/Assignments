```python
'''Task 1 — Create and Serialize a Nested JSON Object Create a Python dictionary representing a student with keys: name, age, passed (boolean), marks (list of three numbers), 
and a nested address object containing city and Pin code. Convert it to a formatted JSON string using json.dumps() with indent=4 and print the result.
'''
import json
Student = {"name" : "Pavan",
           "age" : 26,
           "passed" : True,
           "marks" : [98, 94, 92],
           "address" : {"city" : "Hyderabad",
                        "Pin_code" : 502032}
            }

Student_data = json.dumps(Student, indent = 4)
print("JSON Output :", Student_data)
```

```python
'''Task 2 — Parse JSON and Access Values Using json.loads(), parse the JSON string from Task 1 back into a Python dictionary. 
Print the student's name, the second marks value, and the city from the nested address.'''

JSON_to_Dict = json.loads(Student_data)

StudentName = JSON_to_Dict["name"]
SecondMarks = JSON_to_Dict["marks"][1]
City = JSON_to_Dict["address"]["city"]
print(f" Name : {StudentName}, SecondMarks : {SecondMarks}, City : {City}")
```

```python
'''Task 3 — Identify and Fix Invalid JSON The following JSON string contains two syntax errors. Identify them in a comment, fix the string, parse it using json.loads(), 
and print the result.\n\npython\nbad_json = '{ name: \"Raj\", \"score\": 88, \"active\": true, }'\n'''

# bad_json = { name: "Raj", "score": 88, "active": true, }
# 1. No double quotes for name variable.
# 2. at the last there is an extra comma.
bad_json = '{ "name": "Raj", "score": 88, "active": true}'
parsed = json.loads(bad_json)
print(f"Parsed JSON : {parsed}")
```
