```python
Capacity = 350

total_bookings = 0
tickets_sold = 0
rejected_bookings = 0

while True:
    No_of_tickets = int(input("Enter the number of tickets : "))
    if No_of_tickets == 0:
        break

    if No_of_tickets < 1 or No_of_tickets > 15 or No_of_tickets > Capacity:
        print("Booking rejected. Invalid number of tickets.")
        rejected_bookings += 1
        continue

    if No_of_tickets > Capacity:
        print("Booking rejected. Not enough seats available.")
        rejected_bookings += 1
        continue

    Ages = []
    for i in range(No_of_tickets):
        Age = int(input(f"Enter the age of person {i+1}: "))
        Ages.append(Age)
    
    Valid_Age = True
    for age in Ages:
        if age < 12:
            Valid_Age = False
            break
    if not Valid_Age:
        print("Booking rejected. One or more passengers are not eligible.")
        rejected_bookings += 1
        continue

    Capacity -= No_of_tickets
    tickets_sold += No_of_tickets
    total_bookings += 1

    print(f"Booking successful. Tickets sold: {No_of_tickets}, Remaining Seats: {Capacity}")
    
print(f"Final Report: Total Bookings : {total_bookings}, Total Tickets Sold : {tickets_sold}, Rejected Bookings : {rejected_bookings}, Remaining Seats : {Capacity}")
```
