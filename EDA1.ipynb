```python
import pandas as pd

df = pd.read_csv("titanic.csv")
```

```python
df_cleaned = df[["Pclass","Sex","Survived","Age","Fare"]]
```

```python
summary = df_cleaned.groupby(["Pclass","Sex"]).agg({"Survived" : ["count","sum","mean"],"Age":["mean"],"Fare":"max"}).reset_index()
```

```python
summary.columns = ["Pclass","gender","total_passengers","survivors","survival_rate","avg_age","max_fare"]
summary
```
